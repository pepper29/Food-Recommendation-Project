### voca_dictionary DB 수동 구축

```python
import pandas as pd

voca_dict={}
check=0
start=0
check_index=''
while True:
    print("---------dictionary 만들기-----------")
    print(" 1 : 새로운 key값 만들기             ")
    print(" 2 : key의 해당하는 value 작성       ")
    print(" 3 : key 앞글자 검색                 ")
    print(" 4 : 저장 및 종료                    ")
    print("-------------------------------------")
    if start==0:
        try:
            data=pd.read_csv("voca_dictionary.csv", encoding='utf-8-sig', index_col=0)
            check=1
            key='n'
            for st in data.index:
                check_index=' '+st
            
        except:
            print("Create New Data Settings")
            key='n'
        
    start=1
    try:
        num=int(input("select number : "))

        if num==1:
            key=input("key 입력: ")
            try:
                for st in voca_dict.keys():
                    check_index=' '+st
                if key not in check_index.split():
                    voca_dict[str(key)]=''
            except:
                print('',end='')
            print("\n complete \n")

        elif num==2:
            print("-- 숫자 0 입력 시 value 값 입력 종료 --")
            while True:
                if key=='n':
                    print("\nkey insert plz")
                    break
                value=''
                value=input("value 입력: ")
                if value=='0':
                    print('\n')
                    break
                checking=''
                if check==1:
                    for st in data.values:
                        checking+=' '+st[0]
                    for st in voca_dict.values():
                        checking+=' '+st
                else:
                    for st in voca_dict.values():
                        checking+=' '+st
                for name in value.split():
                    if name not in checking.split():
                        voca_dict[str(key)]+=' '+str(name)
                        checking+=' '+str(name)
                    else:
                        print("Already data:", name)
                print(f'\nEnter the {str(key)} = {voca_dict[str(key)]}')
        elif num==3:
            search=input("앞자리 search: ")
            sear=[]
            if len(search)==1:
                ch=''
                for st in data.index:
                    ch+=' '+st
                for st in voca_dict.keys():
                    ch+=' '+st
                for name in ch.split():
                    if search==name[0]:
                        sear.append(name)
                if len(sear)!=0:
                    print("\n>>>>>>>>>>>>>> Search Keys <<<<<<<<<<<<<<<<<<")
                    print(sear)
                    print('\n')
                else:
                    print("\n== Not Keys ==\n")
            else:
                print('\nSearch error\n')
        elif num==4:
            if check==1:
                for key in voca_dict.keys():
                    if key in data.index:
                        data.loc[str(key),'values']+=' '+str(voca_dict[str(key)])
                    else:
                        temp=pd.DataFrame(voca_dict[str(key)], index=[str(key)], columns=['values'])
                        data=data.append(temp)
                try:
                    data.to_csv("voca_dictionary.csv", encoding='utf-8-sig', index=True)
                    break
                except:
                    voca_dict={}
                    print("\nSave error\n")
            else:
                data=pd.DataFrame(voca_dict.values(),index=voca_dict.keys(), columns=['values'])
                data.to_csv("voca_dictionary.csv", encoding='utf-8-sig', index=True)
                break
        else:
            print('\nCommand Error\n')
    except:
        print('\nCommand Error\n')
```

```python
# 만들어진 파일 중복 검사

checking=''
check_index=[]
for st in data.values:
    checking+=' '+st[0]
for st in data.index:
    checking+=' '+st
    check_index.append(st)
    
for key in data2.index:
    
    if key not in check_index:
        temp=pd.DataFrame(' ',index=[str(key)], columns=['values'])
        data=data.append(temp)
        check_index.append(str(key))
        i=0
        for n in data.values:
            if key in n[0].split():
                data.iloc[i,0]=data.iloc[i,0].replace(" "+str(key)+" "," ")
            i+=1

    values=data2.loc[str(key),'values']
    for name in values.split():
        if name not in checking.split():
            data.loc[str(key),'values']+=' '+str(name)
            checking+=' '+str(name)
        
display(data)
```
