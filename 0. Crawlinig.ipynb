```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import urllib.request
import pandas as pd

# 고정값
rows=10
page_num=10
# search_list = '냉면,전골,곱창,국수,떡볶이'
#함수화
def crawling():

    #웹드라이버창 오픈
    chrome_options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

    # URL접속
    url = 'https://www.mangoplate.com/'
    driver.get(url)

    # 검색어 입력 및 조회
    search_name = '냉면'
    input_box = driver.find_element(By.CSS_SELECTOR, 'input.HomeSearchInput')
    input_box.clear()
    input_box.send_keys(search_name)
    input_box.send_keys(Keys.ENTER)
    time.sleep(3)

    #리뷰들을 모을 리스트 생성
    review_box = []

    #페이지
    for k in range(1, page_num + 1):
        page= f'body > main > article > div.column-wrapper > div > div > section > div.paging-container > p > a:nth-child({k})'
        driver.find_element(By.CSS_SELECTOR, page).click()
        time.sleep(3)
        print('-----------',k,"페이지를 불러옵니다.",'-----------')

        #행
        for j in range(1, rows + 1):
            #j번으로 페이지넘기기
            selc = f'body > main > article > div.column-wrapper > div > div > section > div.search-list-restaurants-inner-wrap > ul > li:nth-child({j})'
            driver.find_element(By.CSS_SELECTOR, selc).click()
            time.sleep(3)
            print(j,'번째행 크롤링을 시작합니다.')

            #열
            for i in range(1, 3):
            #j번째행 i번째 크롤링을 시작
                selector = f'body > main > article > div.column-wrapper > div > div > section > div.search-list-restaurants-inner-wrap > ul > li:nth-child({j}) > div:nth-child({i}) > figure'
                driver.find_element(By.CSS_SELECTOR, selector).click()
                time.sleep(3)
                print(j,'번째행',i,'번째 크롤링을 시작합니다.')

                #해당 아이템의 리뷰수를 확인
                review_num = driver.find_element(By.CSS_SELECTOR, 'body > main > article > div.column-wrapper > div.column-contents > div > section.RestaurantReviewList > header > h2 > span.RestaurantReviewList__AllCount').text
                num = int(review_num)
                print(f'{j}행{i}열 리뷰수:',num)

                #마지막 리뷰가 나올때까지 '더보기' 클릭 &스크롤 내리기
                last_height = driver.execute_script("return document.body.scrollHeight")  # 브라우저의 높이를 자바스크립트로 찾음
                while True:

                    # 스크롤다운
                    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  # 브라우저 끝까지 스크롤을 내림
                    time.sleep(2)

                    # 브라우저가 다 내려갔는지 판단
                    new_height = driver.execute_script("return document.body.scrollHeight")

                    if new_height == last_height: # 높이가 같은지 판단후
                        #하단에 더보기 클릭
                        button = 'body > main > article > div.column-wrapper > div.column-contents > div > section.RestaurantReviewList > div.RestaurantReviewList__MoreReviewButton'
                        driver.find_element(By.CSS_SELECTOR, button).click()
                        time.sleep(5)
                        print('더보기 클릭')

                        while True:
                            if new_height != last_height: #더보기를 누르고 높이가 같지 않으면 continue
                                driver.find_element(By.CSS_SELECTOR, button).click()
                                time.sleep(2)
                            elif new_height == last_height: #더보기를 누르고 높이가같다면 stop
                                print('더이상 불러올 리뷰가없습니다.')
                                break


                last_height = new_height

                # 한페이지에 있는 리뷰를 리스트에 담기
                for n in range(1,num+1):
                    try:
                        reviews = driver.find_element(by=By.XPATH, value= f'/html/body/main/article/div[1]/div[1]/div/section[3]/ul/li[{i}]/a/div[2]/div/p').text
                        review_box.append(reviews)
                    except:
                        continue



                #이전페이지로 돌아감(다음 아이템을 클릭하기 위해)
                driver.back()
                print(j,"행",i,"번째 크롤링완료")

            print(j,"행 크롤링완료")
            
        #데이터 프레임화
        data = pd.DataFrame(review_box)

        print(k,"페이지 크롤링완료")

                    

    #csv파일로 저장하기
    data.to_csv(f'6.{search_name}.csv', header=False, index=False)
    print("저장 완료")
```

## 예전코드

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import urllib.request
import pandas as pd

# 고정값
rows=10
page_num=10
# search_list = '냉면,전골,곱창,국수,떡볶이'

#함수화
def crawling():
    
    #웹드라이버창 오픈
    chrome_options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
    
    # URL접속
    url = 'https://www.mangoplate.com/'
    driver.get(url)
    
    # 검색어 입력 및 조회
    search_name = '간장게장'
    input_box = driver.find_element(By.CSS_SELECTOR, 'input.HomeSearchInput')
    input_box.clear()
    input_box.send_keys(search_name)
    input_box.send_keys(Keys.ENTER)
    time.sleep(1)
    
    #리뷰들을 모을 리스트 생성
    review_box = []
    
    #페이지
    for k in range(1, page_num + 1):
        page= f'body > main > article > div.column-wrapper > div > div > section > div.paging-container > p > a:nth-child({k})'
        driver.find_element(By.CSS_SELECTOR, page).click()
        time.sleep(3)
        print('-----------',k,"페이지를 불러옵니다.",'-----------')
        
        #행
        for j in range(1, rows + 1):
            #j번으로 페이지넘기기
            selc = f'body > main > article > div.column-wrapper > div > div > section > div.search-list-restaurants-inner-wrap > ul > li:nth-child({j})'
            driver.find_element(By.CSS_SELECTOR, selc).click()
            time.sleep(3)
            print(j,'번째행 크롤링을 시작합니다.')
            
            #열
            for i in range(1, 3):
                #j번째행 i번째 크롤링을 시작
                selector = f'body > main > article > div.column-wrapper > div > div > section > div.search-list-restaurants-inner-wrap > ul > li:nth-child({j}) > div:nth-child({i}) > figure'
                driver.find_element(By.CSS_SELECTOR, selector).click()
                time.sleep(3)
                print(j,'번째행',i,'번째 크롤링을 시작합니다.')
                
                #리뷰 수 파악
                review_num = driver.find_element(By.CSS_SELECTOR, 'body > main > article > div.column-wrapper > div.column-contents > div > section.RestaurantReviewList > header > h2 > span.RestaurantReviewList__AllCount').text
                review_num = int(review_num)
                print(f'{j}행{i}열 리뷰수:',review_num)
                
                # 스크롤내리며 더보기 버튼
                #마지막 리뷰가 나올때까지 '더보기' 클릭 &스크롤 내리기
                last_height = driver.execute_script("return document.body.scrollHeight")  # 현재 높이를 알려줌
                while True:
                    # 스크롤다운
                    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  # 브라우저 끝까지 스크롤을 내림
                    time.sleep(3)
                    
                    # 브라우저가 다 내려갔는지 판단
                    new_height = driver.execute_script("return document.body.scrollHeight")    # 다 내려간 곳의 높이 지정
                    
                    if new_height == last_height:
                        try:
                            #하단에 더보기 클릭
                            button = 'body > main > article > div.column-wrapper > div.column-contents > div > section.RestaurantReviewList > div.RestaurantReviewList__MoreReviewButton'
                            driver.find_element(By.CSS_SELECTOR, button).click()
                            time.sleep(5)
                            
                            while True:
                                if new_height == last_height:
                                    break
                                    
                                elif new_height != last_height:
                                    driver.find_element(By.CSS_SELECTOR, button).click()
                                    time.sleep(5)
                        except:
                            break
                            
                    last_height = new_height
                    
                # 한페이지에 있는 리뷰를 리스트에 담기
                cnt = 0
                for n in range(1, review_num+1):
                    try:
                        reviews = driver.find_element(by=By.XPATH, value= f'/html/body/main/article/div[1]/div[1]/div/section[3]/ul/li[{n}]/a/div[2]/div/p').text
                        review_box.append(reviews)
                        cnt += 1 #카운팅하기위해
                        print(k,'p',j,'행',i,'열','리스트에 추가된 개수는',cnt,'입니다.')
                        time.sleep(3)
                    except:
                        continue
                        
                #데이터 프레임화
                data = pd.DataFrame(review_box)
                
                #이전페이지로 돌아감(다음 아이템을 클릭하기 위해)
                driver.back()
                
                print(j,"행",i,"번째 크롤링완료")
                
            print(j,"행 크롤링완료")
            
        print(k,"페이지 크롤링완료")
        
    #csv파일로 저장하기
    data.to_csv(f'6.{search_name}.csv', header=False, index=False)
    print("저장 완료")
```
